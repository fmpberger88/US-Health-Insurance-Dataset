Import U.S Medical Insurance Costs Dataset
https://www.kaggle.com/datasets/teertha/ushealthinsurancedataset

```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('insurance.csv')
```

Descriptive statistic analysis

```python
print(data.describe(include='all'))
```

```python
mean_age = np.mean(data.age)
std_age = np.std(data.age)
print(mean_age)
print(std_age)


plt.hist(data.age, bins=50, density=False)
plt.title(r'Histogram of Age: $\mu=39.21$, $\sigma=14.01$')
plt.suptitle('US Health Insurance Dataset, kaggle.com')
plt.xlabel('Age in Years')
plt.ylabel('Number of Insured Persons')
plt.axvline(data.age.mean(), color='k', linestyle='dashed', linewidth=1)
min_ylim, max_ylim = plt.ylim()
min_xlim, max_xlim = plt.xlim()
plt.text(data.age.mean(), max_ylim*0.9, 'Mean: {:.2f}'.format(data.age.mean()))
plt.show()
```

```python
mean_bmi = np.mean(data.bmi)
std_bmi = np.std(data.bmi)
print(mean_bmi)
print(std_bmi)

plt.hist(data.bmi, bins=50, density=False)
plt.title(r'Histogram of BMI: $\mu=33.66$, $\sigma=6.10$')
plt.suptitle('US Health Insurance Dataset, kaggle.com')
plt.xlabel('Body Mass Index (BMI)')
plt.ylabel('Number of Insured Persons')
plt.axvline(data.bmi.mean(), color='k', linestyle='dashed', linewidth=1)
min_ylim, max_ylim = plt.ylim()
min_xlim, max_xlim = plt.xlim()
plt.text(data.bmi.mean(), max_ylim*0.9, 'Mean: {:.2f}'.format(data.bmi.mean()))
plt.show()

```

```python
mean_charges = np.mean(data.charges)
std_charges = np.std(data.charges)
print(mean_charges)
print(std_charges)

plt.hist(data.charges, bins=50, density=False)
plt.title(r'Histogram of Insurance Cost: $\mu=33.66$, $\sigma=6.10$')
plt.suptitle('US Health Insurance Dataset, kaggle.com')
plt.xlabel('Insurance Cost per Year')
plt.ylabel('Number of Insured Persons')
plt.axvline(data.charges.mean(), color='k', linestyle='dashed', linewidth=1)
min_ylim, max_ylim = plt.ylim()
min_xlim, max_xlim = plt.xlim()
plt.text(data.charges.mean(), max_ylim*0.9, 'Mean: {:.2f}'.format(data.charges.mean()))
plt.show()
```
